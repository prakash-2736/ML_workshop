```python
print(f"Number of duplicate rows: {googleStocksprice.duplicated().sum()}")


googleStocksprice = googleStocksprice.drop_duplicates()
```

```python

googleStocksprice['Close'].hist(bins=30, figsize=(10, 6))
plt.title("Distribution of Closing Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(y=googleStocksprice['Close'])
plt.title("Boxplot of Closing Prices")
plt.show()
```

```python
googleStocksprice['Close'].plot(figsize=(14, 7))
plt.title("Google Stock Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
correlation_matrix = googleStocksprice.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.scatterplot(x=googleStocksprice['Open'], y=googleStocksprice['Close'])
plt.title("Opening vs Closing Prices")
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.show()
```
